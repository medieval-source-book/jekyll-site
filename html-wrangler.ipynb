```python
import os
from html_to_markdown import convert_to_markdown

def convert_html_to_markdown(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".html"):
            html_path = os.path.join(directory, filename)
            markdown_path = os.path.splitext(html_path)[0] + ".md"
            with open(html_path, "r") as html_file:
                html_content = html_file.read()
            markdown_content = convert_to_markdown(html_content)
            with open(markdown_path, "w") as markdown_file:
                markdown_file.write(markdown_content)
            print(f"Converted {html_path} to {markdown_path}")

# Usage
convert_html_to_markdown("html")
```
